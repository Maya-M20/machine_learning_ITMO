$1.$ Определить (с обоснованием), зависимы или независимы следующие события:

a) Несовместные события

События А и В несовместны, если $A\cap B = \emptyset$
Независимость - $P(A|B) = P(A) $

$P(A|B) = \frac{P(A\cap B)}{P(B)} = \frac{P(AB)}{P(B)} = 0 \neq P(A)$

Следовательно, события независимы.

b) События, образующие σ-алгебру Σ в пространстве (Ω,Σ,P)

Могут быть как зависимые, так и независимые:

1) при $A\cap B \neq \emptyset$ - зависимые
2) при $A\cup B - A\cap B \neq \emptyset$ - независимые

c) События, имеющие одинаковую вероятность $P(A) = P(B)$

Может быть два случая:
1) $P(A) = P(B)$ и события зависимы (на одной монете: А - выпал орел, В - выпала решка)
2) $P(A) = P(B)$ и события независимы (А - на первой монете выпал орел, В - на второй выпала решка)

$4.$ Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того, что одному из них придется ожидать освобождения причала,
если время стоянки первого парохода — 1 час, а второго — 2 часа.


$Sполосы = 24^2 - (\frac{23^2}{2} + \frac{22^2}{2}) = \frac{24^2 - 23^2}{2} + \frac{24^2 - 22^2}{2} = \frac{24^2 - (24 - 1)^2 + 24^2 - (24 - 2)^2}{2} = \frac{48 - 1  + 96 - 4}{2} = \frac{139}{2}$


$P(ожидания) = \frac{139}{2*24^2} = \frac{139}{1152}$


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

plt.figure(figsize=(10, 8))

x = np.linspace(0, 24, 400)
y = np.linspace(0, 24, 400)

plt.plot(x, x + 1, 'r-', linewidth=2, label='y = x + 1 (Второй ждет первого)')
plt.plot(x, x - 2, 'b-', linewidth=2, label='y = x - 2 (Первый ждет второго)')

x_fill = np.linspace(0, 24, 100)
y_upper = x_fill + 1
y_lower = x_fill - 2

y_upper = np.clip(y_upper, 0, 24)
y_lower = np.clip(y_lower, 0, 24)

vertices = []
for i in range(len(x_fill)):
    vertices.append([x_fill[i], y_upper[i]])
for i in range(len(x_fill)-1, -1, -1):
    vertices.append([x_fill[i], y_lower[i]])

vertices = np.array(vertices)
polygon = Polygon(vertices, alpha=0.3, color='green', label='Ожидание')
plt.gca().add_patch(polygon)

plt.xlabel('Время прибытия 1го парохода (х, часы)', fontsize=12)
plt.ylabel('Время прибытия 2-го парохода (у, часы)', fontsize=12)

plt.xlim(0, 24)
plt.ylim(0, 24)
plt.grid(True, alpha=0.4)

plt.legend(loc='upper left')

plt.gca().set_aspect('equal')

plt.tight_layout()
plt.show()
```

дании в самолет одного снаряда, снаряд с вероятностью p1 попадает в
первую часть, с вероятностью p2 — во вторую, с вероятностью p3 — в
третью. Попавшие снаряды распределяются по частям независимо друг
от друга.

ность P(A|m) события A – «Самолет поражен» – при m = 1, 2, 3, 4

$P(A|m=1) = p1$

$P(A|m=2) = p1 + p2$

$P(A|m=4) = p1*p2 + p1*p3 + p2 + p3$
