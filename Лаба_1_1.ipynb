##1.1 Вероятностное пространство, формула Байеса

$1.$

a) Несовместные события

События А и В несовместны, только если $P(AB) = 0$
Независимость - $P(A|B) = P(A) $

$P(A|B) = \frac{P(A\cap B)}{P(B)} = \frac{P(AB)}{P(B)} = 0 \neq P(A)$

Следовательно, события независимы.

b) -

c) События, имеющие одинаковую вероятность $P(A) = P(B)$

Может быть два случая:
1) $P(A) = P(B)$ и события зависимы (на одной монете: А - выпал орел, В - выпала решка)
2) $P(A) = P(B)$ и события независимы (А - на первой монете выпал орел, В - на второй выпала решка)

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

plt.figure(figsize=(10, 8))

x = np.linspace(0, 6, 400)
y = np.linspace(0, 6, 400)

plt.plot(x, x + 1, 'r-', linewidth=2, label='y = x + 1 (Второй ждет первого)')
plt.plot(x, x - 2, 'b-', linewidth=2, label='y = x - 2 (Первый ждет второго)')

x_fill = np.linspace(0, 6, 100)
y_upper = x_fill + 1
y_lower = x_fill - 2

y_upper = np.clip(y_upper, 0, 6)
y_lower = np.clip(y_lower, 0, 6)

vertices = []
for i in range(len(x_fill)):
    vertices.append([x_fill[i], y_upper[i]])
for i in range(len(x_fill)-1, -1, -1):
    vertices.append([x_fill[i], y_lower[i]])

vertices = np.array(vertices)
polygon = Polygon(vertices, alpha=0.3, color='green', label='Ожидание')
plt.gca().add_patch(polygon)

plt.xlabel('Время прибытия 2го парохода (х, часы)', fontsize=12)
plt.ylabel('Время прибытия 1-го парохода (у, часы)', fontsize=12)

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid(True, alpha=0.4)

plt.legend(loc='upper left')

plt.gca().set_aspect('equal')

plt.tight_layout()
plt.show()
```
