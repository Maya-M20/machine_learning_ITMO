# Задача 1

Пусть $X_1, X_2, \ldots, X_n$ — выборка из экспоненциального распределения с параметром $\lambda$. Найти оценку максимального правдоподобия параметра $\lambda$, сравнить ее с байесовской оценкой (MAP и математическое ожидание апостреорного распределения), подобрав сопряженное распределение. Сравнить полученные байесовские оценки с оценкой MLE. Найти предсказательное распределение

Плотность выборки

$$
f(x \mid \lambda) = \lambda e^{-\lambda x}, \quad x \ge 0,\ \lambda > 0.
$$

### MLE:

$$
L(\lambda) = \prod_{i=1}^{n} \lambda e^{-\lambda X_i}
= \lambda^{n} e^{-\lambda \sum_{i=1}^{n} X_i}.
$$


$$
\ell(\lambda) = n \ln \lambda - \lambda \sum_{i=1}^{n} X_i.
$$


Возьмём производную по $\lambda$ и приравняем к нулю:

$$
\frac{\partial \ell}{\partial \lambda}
= \frac{n}{\lambda} - \sum_{i=1}^{n} X_i = 0.
$$


$$
\hat{\lambda}*{\text{MLE}}
= \frac{n}{\sum*{i=1}^{n} X_i}
= \frac{1}{\overline{X}}.
$$


### Байесовская оценка (сопряжённое априорное распределение)

Экспоненциальное распределение входит в экспоненциальное семейство в форме

$$
f(x \mid \lambda) = \lambda e^{-\lambda x}
= e^{\ln \lambda - \lambda x},
$$

где параметр — $\lambda$.

Для параметра масштаба $\lambda$ сопряжённым априорным распределением является гамма-распределение
$\mathrm{Gamma}(\alpha, \beta)$ с плотностью

$$
p(\lambda)
= \frac{\beta^{\alpha}}{\Gamma(\alpha)}
, \lambda^{\alpha - 1} e^{-\beta \lambda},
\quad \lambda > 0.
$$

### Апостериорное распределение

По теореме Байеса:

$$
p(\lambda \mid X_1, \ldots, X_n)
\propto L(\lambda), p(\lambda)
\propto
\bigl[ \lambda^{n} e^{-\lambda n \overline{X}} \bigr]
\cdot
\bigl[ \lambda^{\alpha - 1} e^{-\beta \lambda} \bigr].
$$

Объединяя степени, получаем:

$$
p(\lambda \mid \mathbf{X})
\propto
\lambda^{n + \alpha - 1}
e^{-\lambda (n \overline{X} + \beta)}.
$$

Это форма плотности гамма-распределения:

$$
\lambda \mid \mathbf{X} \sim \mathrm{Gamma}(\alpha', \beta'),
$$

где

$$
\alpha' = n + \alpha,
$$

$$
\beta' = n \overline{X} + \beta.
$$

### MAP:


Мода гамма-распределения $\text{Gamma}(\alpha', \beta')$ при $\alpha' \geq 1$:

$$
\frac{\alpha' - 1}{\beta'} = \frac{n + \alpha - 1}{n\bar{X} + \beta}.
$$


$$
\hat{\lambda}_{\text{MAP}} = \frac{n + \alpha - 1}{n\bar{X} + \beta}.
$$

### Предсказательное распределение:
Предсказательное распределение для нового наблюдения $X_{n+1}$:

$$
p(x_{\text{new}} | \mathbf{X}) = \int_0^\infty f(x_{\text{new}} | \lambda) p(\lambda | \mathbf{X}) \, d\lambda.
$$

Подставляем:

$$
f(x_{\text{new}} | \lambda) = \lambda e^{-\lambda x_{\text{new}}}, \quad p(\lambda | \mathbf{X}) = \frac{(\beta')^{\alpha'}}{\Gamma(\alpha')} \lambda^{\alpha' - 1} e^{-\beta' \lambda}.
$$

Интеграл:

$$
p(x_{\text{new}} | \mathbf{X}) = \frac{(\beta')^{\alpha'}}{\Gamma(\alpha')} \int_0^\infty \lambda^{\alpha'} e^{-(\beta' + x_{\text{new}})\lambda} \, d\lambda.
$$

Интеграл равен $\Gamma(\alpha' + 1) / (\beta' + x_{\text{new}})^{\alpha' + 1}$.

Учитывая $\Gamma(\alpha' + 1) = \alpha' \Gamma(\alpha')$:

$$
p(x_{\text{new}} | \mathbf{X}) = \frac{(\beta')^{\alpha'}}{\Gamma(\alpha')} \cdot \frac{\Gamma(\alpha' + 1)}{(\beta' + x_{\text{new}})^{\alpha' + 1}} = \alpha' \frac{(\beta')^{\alpha'}}{(\beta' + x_{\text{new}})^{\alpha' + 1}}.
$$

Подставим $\alpha' = n + \alpha$, $\beta' = n\bar{X} + \beta$:

$$
p(x_{\text{new}} | \mathbf{X}) = \frac{n + \alpha}{n\bar{X} + \beta} \left( \frac{n\bar{X} + \beta}{n\bar{X} + \beta + x_{\text{new}}} \right)^{n + \alpha + 1}
$$


$$
{p(x_{\text{new}} | \mathbf{X}) = \frac{n + \alpha}{n\bar{X} + \beta} \left( 1 + \frac{x_{\text{new}}}{n\bar{X} + \beta} \right)^{-(n + \alpha + 1)}}
$$

для $x_{\text{new}} \ge 0$

# Задача 2

**Мультиномиальное распределение**

Пусть проводится серия из $n$ испытаний и в результате каждого испытания происходит ровно одно событие из набора $A_1, A_2, \dots, A_m$, причем вероятности этих событий равны соответственно $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, причем
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

Тогда совместное распределение величин $X_1, X_2, \dots, X_m$, где $X_k$ — число наступлений события $A_k$ в серии из $n$ испытаний, задается вероятностями

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

где $n_1, n_2, \dots, n_m$ — произвольный набор целых неотрицательных чисел, таких что

$$\sum_{i=1}^m n_i = n.$$

Произведите байесовский вывод для мультиномиального распределения: найдите апостериорное распределение, используя в качестве сопоряженного распределения к правдоподобию [распределение Дирихле](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), найдите предсказательное распределение.
